```python
import pandas as pd
accesories = pd.read_excel('./trait-effects/szn1/accesories.xlsx')
background = pd.read_excel('./trait-effects/szn1/background.xlsx')
body = pd.read_excel('./trait-effects/szn1/body.xlsx')
head = pd.read_excel('./trait-effects/szn1/head.xlsx')
headwear = pd.read_excel('./trait-effects/szn1/headwear.xlsx')
mouth = pd.read_excel('./trait-effects/szn1/mouth.xlsx')
special = pd.read_excel('./trait-effects/szn1/special.xlsx')
```

```python
# Set game mode effects for Head
#CACNEA OPTIMIZABLE
def setgamemode(trait, trait_df):
    for i, row in trait.iterrows():
        if 'PVP' in row['IN-GAME APPLICATION']:
            trait_df.at[i, 'PVP'] = row['IN-GAME APPLICATION'].split(': ')[1]
        elif 'PVE' in row['IN-GAME APPLICATION']:
            trait_df.at[i, 'PVE'] = row['IN-GAME APPLICATION'].split(': ')[1]
        elif 'MINING' in row['IN-GAME APPLICATION']:
            trait_df.at[i, 'MINING'] = row['IN-GAME APPLICATION'].split(': ')[1]
        elif 'HOMEBASE' in row['IN-GAME APPLICATION']:
            trait_df.at[i, 'HOMEBASE'] = row['IN-GAME APPLICATION'].split(': ')[1]
        elif 'BOSS BATTLE' in row['IN-GAME APPLICATION']:
            trait_df.at[i, 'BOSS BATTLE'] = row['IN-GAME APPLICATION'].split(': ')[1]
        elif 'BATTLE ROYALE' in row['IN-GAME APPLICATION']:
            trait_df.at[i, 'BATTLE ROYALE'] = row['IN-GAME APPLICATION'].split(': ')[1]
```

```python
accesories_df = pd.DataFrame({
    'Trait': accesories['Accesories'],
    'Value': 'Accesories',
    'Total': accesories['TOTAL #'],
    'Rarity': accesories['% RARITY']
})
```

```python
background_df = pd.DataFrame({
    'Trait': background['Background'],
    'Value': 'Background',
    'Total': background['TOTAL #'],
    'Rarity': accesories['% RARITY']
})
```

```python
body_df = pd.DataFrame({
    'Trait': body['Body'],
    'Value': 'Body',
    'Total': body['TOTAL #'],
    'Rarity': accesories['% RARITY']
})
```

```python
head_df = pd.DataFrame({
    'Trait': head['Head'],
    'Value': 'Head',
    'Total': head['TOTAL #'],
    'Rarity': accesories['% RARITY']
})
```

```python
headwear_df = pd.DataFrame({
    'Trait': headwear['Eyes:Headwear:Eyewear'],
    'Value': 'Eyes:Headwear:Eyewear',
    'Total': headwear['TOTAL #'],
    'Rarity': accesories['% RARITY']
})
```

```python
mouth_df = pd.DataFrame({
    'Trait': mouth['Mouth'],
    'Value': 'Mouth',
    'Total': mouth['TOTAL #'],
    'Rarity': accesories['% RARITY']
})
```

```python
special_df = pd.DataFrame({
    'Trait': special['Special'],
    'Value': 'Special',
    'Total': special['TOTAL #'],
    'Rarity': accesories['% RARITY']
})
```

```python
setgamemode(accesories, accesories_df)
setgamemode(background, background_df)
setgamemode(body, body_df)
setgamemode(head, head_df)
setgamemode(headwear, headwear_df)
setgamemode(mouth, mouth_df)
#setgamemode(special, special_df)
```

```python
df = pd.concat([accesories_df, background_df, body_df, head_df, headwear_df, mouth_df], ignore_index=True)
df
```

```python
df.to_csv('in-game-effects.csv')
```
